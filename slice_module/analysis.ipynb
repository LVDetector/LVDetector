```python
import os,json
import pandas as pd
import copy
excel_path = '/home/sink_main/Teamwork_VulnerabilityAnalysis/slice_module/ground_truth_0.xlsx'
path = './slice_result/'
result = {}
for dir in os.listdir(path):
    dir_path = os.path.join(path,dir)
    files = os.listdir(dir_path)
    is_latent = 0
    if 'feature_dict.json' in files:
        with open(os.path.join(dir_path,'feature_dict.json'),'r') as feature_file:
            feature_data = json.load(feature_file)
            for key,value in list(feature_data.values())[0].items():
                if value['label'] == 'Latent_vulnerability':
                     is_latent = 1
                     break
    result[dir+'.c'] = is_latent

df = pd.read_excel(excel_path)
new_df = copy.deepcopy(df)
for index,row in enumerate(df.values):
    file_name = row[0]
    if file_name in result.keys():
        new_df.at[index,'our_method']= result[file_name]
        print(df.loc[index])
new_df.to_excel('./LVFinder_0.xlsx')
```
